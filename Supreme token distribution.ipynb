```python
# library
import matplotlib.pyplot as plt
from palettable.colorbrewer.qualitative import Pastel1_7
 
# create data
names = ['Team and advisors', 'Marketing', 'Ecosystem Growth and Development', 'Treasury','Airdrop','Token Presale','Supreme Incentivization Program','Liquidity Mining']
size = [10,10,10,15,5,10,15,25]
 
# Create a circle at the center of the plot
my_circle = plt.Circle( (0,0), 0.7, color='white')

from palettable.colorbrewer.qualitative import Pastel1_7
plt.pie(size, labels=names, colors=Pastel1_7.hex_colors)
p = plt.gcf()
p.gca().add_artist(my_circle)

# Show the graph
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

data = [10,10,10,15,5,10,15,25]
labels = ['Team and advisors', 'Marketing', 'Ecosystem Growth and Development', 'Treasury','Airdrop','Token Presale','Supreme Incentivization Program','Liquidity Mining']
colors = sns.color_palette('bright')
ax = plt.pie(data, labels=labels, labeldistance=1.15, wedgeprops = { 'linewidth' : 3, 'edgecolor' : 'white' }, colors = Pastel1_7.hex_colors, autopct = '%0.0f%%')
sns.set(rc = {'figure.figsize':(30,8)})
plt.show()
```
